```python
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# API configuration
# Replace with your actual, valid OpenWeatherMap API key
api_key = "b896867fadef710786c586193599616d"
city = "London"  # Example city, change as needed
url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"

# Fetch data
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    data = response.json()

    # Extract relevant data (example: temperature, humidity, wind speed)
    temp = data["main"]["temp"]
    humidity = data["main"]["humidity"]
    wind_speed = data["wind"]["speed"]

    # Example: Prepare data for visualization (bar plot)
    categories = ['Temperature (Â°C)', 'Humidity (%)', 'Wind Speed (km/h)']
    values = [temp, humidity, wind_speed]

    # Create a bar plot
    plt.figure(figsize=(8, 6))
    sns.barplot(x=categories, y=values, palette="viridis")
    plt.title(f"Weather Data for {city}")
    plt.ylabel("Value")
    plt.show()

    # Save as image (optional)
    plt.savefig('weather_data.png')
else:
    # Print an informative error message if the request failed
    print(f"Error fetching weather data. Status code: {response.status_code}")
    print(f"Response message: {response.json().get('message', 'No error message provided')}")
    print("Please check your API key and city name.")
```
